```python
class ROI():
    def __init__(self):
        self.dict = {
            'income': {
                'rental_income': 0,
                'laundry': 0,
                'storage': 0,
                'misc': 0, 
                'total': 0
            },

            'expenses': {
                'tax': 0, 
                'insurance': 0,
                'utilities': {
                    'electic': 0,
                    'water': 0,
                    'sewer': 0,
                    'garbage': 0,
                    'gas': 0
                },
                'hoa': 0,
                'lawn/snow': 0,
                'vacancy': 0,
                'repairs': 0,
                'capex': 0,
                'property_management': 0,
                'mortgage': 0, 
                'total': 0
            },

            'cash_flow': {
                'cash_flow_total': 0
            },

            'roi': {
                'down_payment': 0,
                'closing_costs': 0,
                'rehab_budget': 0,
                'misc': 0,
                'total_investment': 0, 
                'annual_cash_flow': 0, 
                'cash_on_cash_roi': 0
            }
        }

    def incomeMethod(self):
        income_payments = []

        user_income = int(input("What are you charging for rental income? "))
        income_payments.append(user_income)
        self.dict['income']['rental_income'] = user_income 

        user_laundry = int(input("What are you charging for laundry? "))
        income_payments.append(user_laundry)
        self.dict['income']['laundry'] = user_laundry

        user_storage = int(input("What are you charging for storage? "))
        income_payments.append(user_storage)
        self.dict['income']['storage'] = user_storage

        user_misc = int(input("What are you charging for any miscellaneous items? Enter a total number. "))
        income_payments.append(user_misc)
        self.dict['income']['misc'] = user_misc

        user_income_total = sum(income_payments)
        self.dict['income']['total'] = user_income_total
        print(user_income_total)
        print(self.dict)


    def expensesMethod(self):
        expense_payments = []

        user_tax = int(input("What are your monthly taxes? "))
        expense_payments.append(user_tax)
        self.dict['expenses']['tax'] = user_tax

        user_insurance = int(input("What do you pay monthly for insurance? "))
        expense_payments.append(user_insurance)
        self.dict['expenses']['insurance'] = user_insurance

        user_utilities_electric = int(input("What do you pay monthly for electic? "))
        expense_payments.append(user_utilities_electric)
        self.dict['expenses']['utilities']['electic'] = user_utilities_electric

        user_utilities_water = int(input("What do you pay monthly for water? "))
        expense_payments.append(user_utilities_water)
        self.dict['expenses']['utilities']['water'] = user_utilities_water

        user_utilities_sewer = int(input("What do you pay monthly for sewer? "))
        expense_payments.append(user_utilities_sewer)
        self.dict['expenses']['utilities']['sewer'] = user_utilities_sewer

        user_utilities_garbage = int(input("What do you pay monthly for garbage? "))
        expense_payments.append(user_utilities_garbage)
        self.dict['expenses']['utilities']['garbage'] = user_utilities_garbage

        user_utilities_gas = int(input("What do you pay monthly for gas? "))
        expense_payments.append(user_utilities_gas)
        self.dict['expenses']['utilities']['gas'] = user_utilities_gas

        user_hoa = int(input("What do you pay monthly for HOA fees? "))
        expense_payments.append(user_hoa)
        self.dict['expenses']['hoa'] = user_hoa

        user_lawn_snow = int(input("What do you pay monthly for lawn/snow removal? "))
        expense_payments.append(user_lawn_snow)
        self.dict['expenses']['lawn/snow'] = user_lawn_snow

        user_vacancy = int(input("What do you pay monthly for vacancy insurance? "))
        expense_payments.append(user_vacancy)
        self.dict['expenses']['vacancy'] = user_vacancy

        user_repairs = int(input("What do you pay monthly for repairs for you property? "))
        expense_payments.append(user_repairs)
        self.dict['expenses']['repairs'] = user_repairs

        user_capex = int(input("What do you pay monthly for capex? "))
        expense_payments.append(user_capex)
        self.dict['expenses']['capex'] = user_capex

        user_property_management = int(input("What do you pay monthly for property management? "))
        expense_payments.append(user_property_management)
        self.dict['expenses']['property_management'] = user_property_management

        user_mortgage = int(input("What do you pay monthly for your mortgage? "))
        expense_payments.append(user_mortgage)
        self.dict['expenses']['mortgage'] = user_mortgage

        user_expense_total = sum(expense_payments)
        self.dict['expenses']['total'] = user_expense_total
        print(expense_payments)
        print(user_expense_total)
        print(self.dict)

    def cash_flow(self):
        total_income = self.dict['income']['total']
        print(self.dict['income']['total'])
        total_expenses = self.dict['expenses']['total']
        print(self.dict['expenses']['total'])
        total_cash_flow = total_income - total_expenses
        self.dict['cash_flow']['cash_flow_total'] = total_cash_flow
        print(self.dict['cash_flow']['cash_flow_total'])


    def cash_on_cash(self):
        cash_on_cash_payments = []

        user_down_payment = int(input("What was your down payment? "))
        cash_on_cash_payments.append(user_down_payment)
        self.dict['roi']['down_payment'] = user_down_payment 
        
        user_closing_costs = int(input("What was your closing costs? "))
        cash_on_cash_payments.append(user_closing_costs)
        self.dict['roi']['closing_costs'] = user_closing_costs 
        
        user_rehab_budget = int(input("What was your rehab budget? "))
        cash_on_cash_payments.append(user_rehab_budget)
        self.dict['roi']['rehab_budget'] = user_rehab_budget 
        
        user_misc = int(input("What was your miscellaneous payments? "))
        cash_on_cash_payments.append(user_misc)
        self.dict['roi']['misc'] = user_misc 
        
        total_investment = (self.dict['roi']['down_payment']) + (self.dict['roi']['closing_costs']) + (self.dict['roi']['rehab_budget']) + (self.dict['roi']['misc'])
        self.dict['roi']['total_investment'] = total_investment
        print(self.dict['roi']['total_investment'])

        annual_cash_flow = (self.dict['cash_flow']['cash_flow_total']) * 12
        self.dict['roi']['annual_cash_flow'] = annual_cash_flow
        print(self.dict['roi']['annual_cash_flow'])

        cash_on_cash_roi = ((self.dict['roi']['annual_cash_flow']) / (self.dict['roi']['total_investment'])) * 100
        self.dict['roi']['cash_on_cash_roi'] = cash_on_cash_roi

        print(self.dict)
        print(str(self.dict['roi']['cash_on_cash_roi']) + "%" + " return on investment")

    def runner(self):
        self.incomeMethod()
        self.expensesMethod()
        self.cash_flow()
        self.cash_on_cash()




test = ROI()
test.runner()
```
